## 2. The question: decision, action, cost of a wrong call

My research question is: **\"Can we use historical search and engagement metrics to prioritize which content items should be reviewed first for refresh/updates to mitigate search visibility and traffic decline?\"**

Here is the framing of this question:
1. **What decision does this improve?** 
   It improves the prioritization decision of the content editorial team: *\"Which specific page should we review, update, or expand today?\"* Instead of choosing pages randomly or based on ad-hoc requests, the team makes data-driven decisions to refresh pages that have high demand but declining visibility.
2. **Who acts on the output, and what do they do?** 
   A **content editor or SEO manager** acts on the output. They will take a prioritized ranked queue of content candidates, look at the provided reason codes (e.g., stale high-traffic page, declining visibility with search demand, low engagement rate compared to position tier), review the content, and update it (e.g., refreshing facts, adding sections, rewriting meta titles/descriptions).
3. **What does a wrong answer cost?** 
   - *False Positive (recommending a page that doesn't need a refresh)*: Wastes editorial hours. Time spent updating a page that doesn't improve represents a direct labor cost and lost opportunity to fix other pages.
   - *False Negative (missing a page that is rapidly decaying)*: Results in a sustained loss of organic clicks, impressions, and user sessions. If the decay is not caught early, search engines may rank competitors higher, making recovery more difficult and expensive.
4. **Why does data or ML help at all?** 
   A simple manual inspection or plain rule-based system is inadequate because the data is highly multi-dimensional and noisy. Visibility and engagement signals (impressions, clicks, average position, scroll rate, content age, intent) are tangled and change dynamically. A learned model can identify complex patterns and non-linear interactions across these signals to predict the probability of decline, ranking the pages more accurately than hand-tuned heuristics.

## 3. Quick look at the data (2-3 real numbers)

To verify the size of the opportunity for Lane 2, I analyzed the starter dataset `data/raw/content_refresh_anonymized.csv`. Here are the three key findings that show why prioritizing content updates is highly valuable:

1. **High Rate of Decline**: **16,262 out of 30,000 pages (54.21%)** in the starter dataset are currently in a declining traffic trend (`trend_direction == 'down'`).
2. **Massive Volume of Search Impressions at Risk**: These declining pages account for **79,994,363 GSC impressions** out of 156,010,989 total GSC impressions in the last 90 days. This means **51.27% of all organic visibility** is associated with pages that are actively losing traction.
3. **Significant Staleness in Decaying Content**: Among the 16,262 declining pages, **5,686 pages** (5,604 in `91-180` days and 82 in `181+` days) have not been updated for more than 3 months. This indicates a large inventory of stale pages where a refresh could potentially reverse or stabilize the decline.

```python
import pandas as pd
import numpy as np

# Load the starter dataset
df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')

# 1. Total and declining pages
total_pages = len(df)
declining_df = df[df['trend_direction'] == 'down']
num_declining = len(declining_df)
pct_declining = (num_declining / total_pages) * 100

print(f"Total pages: {total_pages:,}")
print(f"Declining pages (trend_direction == 'down'): {num_declining:,} ({pct_declining:.2f}%)")

# 2. Impressions at risk
total_impressions = df['impressions_90d'].sum()
declining_impressions = declining_df['impressions_90d'].sum()
pct_impressions_at_risk = (declining_impressions / total_impressions) * 100

print(f"Total GSC impressions (90d): {total_impressions:,}")
print(f"Impressions of declining pages: {declining_impressions:,} ({pct_impressions_at_risk:.2f}%)")

# 3. Staleness among declining pages
stale_declining = declining_df[declining_df['freshness_tier'].isin(['91-180', '181+'])]
num_stale_declining = len(stale_declining)
pct_stale_declining = (num_stale_declining / num_declining) * 100

print(f"Stale declining pages (>90 days since update): {num_stale_declining:,} ({pct_stale_declining:.2f}% of declining pages)")
```

## 4. Careful words: what I can and can't claim

Understanding the boundaries of our conclusions prevents us from making false promises or dangerous assumptions. 

### What my work CAN claim:
- **Decision-Support**: Our model and prioritized list are designed purely for decision-support, helping content managers prioritize their limited hours based on historical evidence.
- **Observed Correlations**: We can identify and quantify observed correlations between page characteristics (e.g., freshness, position, intent) and traffic trends.
- **Directional Signals**: We can claim directional probability (e.g., \"stale pages with low click-through rates are directionally more likely to continue declining than fresh ones\").

### What my work CANNOT claim:
- **Causal Proof**: We cannot claim that a content refresh *causes* recovery. Because this is observational data, we cannot prove that an edit will reverse decay without a randomized controlled experiment (A/B test).
- **\"Predicting Google's Algorithm\"**: We are not reverse-engineering or predicting Google's internal search algorithm. We are merely modeling historical search outcomes.
- **Absolute Guarantees**: We cannot guarantee that any specific page in the prioritized review list will recover its traffic after being refreshed.
